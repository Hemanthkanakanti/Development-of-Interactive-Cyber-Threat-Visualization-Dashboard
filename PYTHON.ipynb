### 1. Variables

A variable is a named storage location in a computer program that holds a value. In Python, you don't need to declare the type of a variable; it's dynamically typed, meaning the type is inferred when you assign a value to it.

```python
# Assigning values to variables
my_integer = 10
my_float = 3.14
my_string = "Hello, Python!"
my_boolean = True

print(f"my_integer: {my_integer}")
print(f"my_float: {my_float}")
print(f"my_string: {my_string}")
print(f"my_boolean: {my_boolean}")
```

### 2. Data Types

Python has several built-in data types. Here are some of the most common ones:

*   **Numeric Types**: `int` (integers), `float` (floating-point numbers), `complex` (complex numbers).
*   **Boolean Type**: `bool` (True or False).
*   **Sequence Types**: `str` (strings), `list`, `tuple`.
*   **Set Types**: `set`, `frozenset`.
*   **Mapping Type**: `dict` (dictionaries).

```python
# Checking data types using type()
print(f"Type of my_integer: {type(my_integer)}")
print(f"Type of my_float: {type(my_float)}")
print(f"Type of my_string: {type(my_string)}")
print(f"Type of my_boolean: {type(my_boolean)}")
```

### 3. Lists

A `list` is an ordered, mutable (changeable) collection of items. Lists are defined by enclosing elements in square brackets `[]`, separated by commas. Lists can contain items of different data types.

```python
# Creating a list
my_list = [1, "apple", 3.14, True]
print(f"Original list: {my_list}")

# Accessing elements (lists are 0-indexed)
print(f"First element: {my_list[0]}")
print(f"Second element: {my_list[1]}")

# Modifying an element
my_list[1] = "orange"
print(f"Modified list: {my_list}")

# Adding an element
my_list.append(False)
print(f"List after append: {my_list}")

# Removing an element
my_list.remove(1)
print(f"List after remove: {my_list}")
```

### 4. Tuples

A `tuple` is an ordered, *immutable* (unchangeable) collection of items. Tuples are defined by enclosing elements in parentheses `()`, separated by commas. Once a tuple is created, you cannot add, remove, or modify its elements.

```python
# Creating a tuple
my_tuple = (1, "banana", 2.71, False)
print(f"Original tuple: {my_tuple}")

# Accessing elements (tuples are 0-indexed)
print(f"First element: {my_tuple[0]}")
print(f"Second element: {my_tuple[1]}")

# Attempting to modify an element (this will raise an error)
# my_tuple[1] = "grape" # Uncommenting this line will cause a TypeError
```

### 5. Sets

A `set` is an unordered collection of *unique* items. Sets are defined by enclosing elements in curly braces `{}` (or using the `set()` constructor for an empty set). Duplicate elements are automatically removed.

```python
# Creating a set
my_set = {1, 2, 3, 2, 4, 1}
print(f"Original set (duplicates removed): {my_set}")

# Adding an element
my_set.add(5)
print(f"Set after adding 5: {my_set}")

# Removing an element
my_set.remove(2)
print(f"Set after removing 2: {my_set}")

# Set operations
set_a = {1, 2, 3}
set_b = {3, 4, 5}

print(f"Union: {set_a.union(set_b)}")
print(f"Intersection: {set_a.intersection(set_b)}")
```

### 6. Dictionaries

A `dictionary` is an unordered, mutable collection of key-value pairs. Each key must be unique and immutable (e.g., strings, numbers, tuples), while values can be of any data type. Dictionaries are defined by enclosing key-value pairs in curly braces `{}`.

```python
# Creating a dictionary
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
print(f"Original dictionary: {my_dict}")

# Accessing values by key
print(f"Name: {my_dict['name']}")
print(f"Age: {my_dict['age']}")

# Modifying a value
my_dict['age'] = 31
print(f"Modified dictionary: {my_dict}")

# Adding a new key-value pair
my_dict['occupation'] = "Engineer"
print(f"Dictionary after adding occupation: {my_dict}")

# Removing a key-value pair
del my_dict['city']
print(f"Dictionary after removing city: {my_dict}")
```

### 2. Pandas Data Structures: Series and DataFrame

#### Series
A `Series` is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). It's essentially a column in a spreadsheet or a single column of data.

#### DataFrame
A `DataFrame` is a two-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or a SQL table, or a dictionary of Series objects. It is the most commonly used pandas object.

```python
# Creating a Series
s = pd.Series([10, 20, 30, 40, 50])
print("Pandas Series:")
print(s)

# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [50000, 60000, 45000, 75000]
}
df = pd.DataFrame(data)
print("\nPandas DataFrame:")
print(df)
```

#### Inspecting Data

- `head()`: Displays the first N rows of the DataFrame.
- `info()`: Provides a concise summary of the DataFrame, including data types and non-null values.
- `describe()`: Generates descriptive statistics of the DataFrame's numerical columns.

```python
print("DataFrame Info:")
data_df.info()

print("\nDataFrame Description (Numerical Columns):")
display(data_df.describe())
```

#### Selecting Columns and Rows

- **Selecting a single column**: `df['column_name']`
- **Selecting multiple columns**: `df[['col1', 'col2']]`
- **Selecting rows by index (label)**: `df.loc[start_index:end_index]`

#### Filtering Data

You can filter rows based on conditions applied to one or more columns.

```python
print("Employees older than 30:")
display(data_df[data_df['Age'] > 30])

print("\nFemale employees from New York:")
display(data_df[(data_df['Gender'] == 'Female') & (data_df['City'] == 'New York')])
```

#### Handling Missing Values

Real-world data often has missing values (represented as `NaN` or `None`). Pandas provides tools to detect and handle them.

- `isna()`: Returns a boolean DataFrame indicating missing values.
- `dropna()`: Removes rows or columns with missing values.
- `fillna()`: Fills missing values with a specified value or method.

```python
# Introduce some missing values for demonstration
data_with_nan = data_df.copy()
data_with_nan.loc[[1, 5], 'Salary'] = np.nan
data_with_nan.loc[3, 'City'] = np.nan

print("DataFrame with missing values:")
display(data_with_nan)

print("\nMissing values per column:")
print(data_with_nan.isna().sum())

print("\nDataFrame after dropping rows with any missing values:")
display(data_with_nan.dropna())

print("\nDataFrame after filling missing 'Salary' with its mean and 'City' with 'Unknown':")
data_filled = data_with_nan.copy()
data_filled['Salary'] = data_filled['Salary'].fillna(data_filled['Salary'].mean())
data_filled['City'] = data_filled['City'].fillna('Unknown')
display(data_filled)
```

#### Grouping and Aggregation

`groupby()` is one of the most powerful pandas features. It allows you to group data based on one or more columns and then apply an aggregation function (like `sum`, `mean`, `count`, `max`, `min`) to each group.

```python
print("Average salary by Gender:")
display(data_df.groupby('Gender')['Salary'].mean())

print("\nTotal employees and average salary by City:")
display(data_df.groupby('City').agg(Total_Employees=('ID', 'count'), Average_Salary=('Salary', 'mean')))
```

## Basic Object-Oriented Programming (OOP) Concepts in Python

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of 'objects', which can contain data and code: data in the form of fields (attributes), and code in the form of procedures (methods). Python is an object-oriented language.

### 1. Classes and Objects

*   **Class**: A class is a blueprint or a template for creating objects. It defines a set of attributes (data) and methods (functions) that the objects will have.
*   **Object**: An object is an instance of a class. It's a real-world entity that has a state (attributes) and behavior (methods).

```python
# Define a class called 'Dog'
class Dog:
    # Class attribute
    species = "Canis familiaris"

    # The constructor method (__init__)
    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

    # Instance method
    def bark(self):
        return f"{self.name} says Woof!"

    # Another instance method
    def description(self):
        return f"{self.name} is {self.age} years old."

# Create objects (instances) of the Dog class
my_dog = Dog("Buddy", 3)
your_dog = Dog("Lucy", 5)

print(f"My dog's name: {my_dog.name}")
print(f"Your dog's age: {your_dog.age}")
print(f"My dog's species: {my_dog.species}")
print(my_dog.bark())
print(your_dog.description())
```

### 2. Attributes

Attributes are variables that store data associated with an object. They can be:

*   **Instance Attributes**: Unique to each instance of a class (e.g., `name`, `age` in the `Dog` class).
*   **Class Attributes**: Shared by all instances of a class (e.g., `species` in the `Dog` class).

```python
# Accessing attributes
print(f"My dog's name (instance attribute): {my_dog.name}")
print(f"My dog's species (class attribute): {Dog.species}")
print(f"Your dog's species (accessing via instance): {your_dog.species}")
```

### 3. Methods

Methods are functions defined inside a class that perform actions on the object's data. They can be:

*   **Instance Methods**: Operate on the instance's attributes (e.g., `bark()`, `description()`). They always take `self` as their first parameter.
*   **Class Methods**: Operate on class attributes and are defined using the `@classmethod` decorator. They take `cls` (the class itself) as their first parameter.
*   **Static Methods**: Do not operate on instance or class attributes and are defined using the `@staticmethod` decorator. They don't take `self` or `cls`.

```python
class Car:
    wheels = 4 # Class attribute

    def __init__(self, make, model):
        self.make = make
        self.model = model

    def display_info(self): # Instance method
        return f"This is a {self.make} {self.model}."

    @classmethod
    def get_wheels_count(cls): # Class method
        return f"All cars have {cls.wheels} wheels."

    @staticmethod
    def start_engine(): # Static method
        return "Engine started!"

my_car = Car("Toyota", "Camry")
print(my_car.display_info())
print(Car.get_wheels_count()) # Call class method using class name
print(my_car.start_engine()) # Call static method using instance
```

### 4. Inheritance

Inheritance allows a class (child/subclass) to inherit attributes and methods from another class (parent/superclass). This promotes code reusability and establishes an "is-a" relationship.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Cat(Animal):
    def __init__(self, name, breed):
        super().__init__(name) # Call parent class constructor
        self.breed = breed

    def speak(self):
        return f"{self.name} the {self.breed} says Meow!"

class Dog(Animal):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def speak(self):
        return f"{self.name} the dog says Woof!"

my_cat = Cat("Whiskers", "Siamese")
my_dog_inherited = Dog("Buddy", 3)

print(my_cat.speak())
print(my_dog_inherited.speak())
print(f"My cat's breed: {my_cat.breed}")
```

### 5. Polymorphism

Polymorphism means "many forms". In OOP, it allows objects of different classes to be treated as objects of a common type. This is often achieved through method overriding (as seen in `speak()` method above) or through interfaces/abstract classes.

```python
def make_animal_speak(animal):
    return animal.speak()

# Using polymorphism with the inherited objects
print(make_animal_speak(my_cat))
print(make_animal_speak(my_dog_inherited))

# Another example with a different class
class Duck(Animal):
    def speak(self):
        return f"{self.name} the duck says Quack!"

daisy = Duck("Daisy")
print(make_animal_speak(daisy))
```

### 6. Encapsulation

Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit (class). It also involves restricting direct access to some of an object's components, which is typically done by using private attributes and methods.

In Python, encapsulation is mostly by convention: a single underscore `_` prefixes a variable or method name to indicate that it's intended for internal use, and a double underscore `__` (name mangling) makes it harder to access from outside the class.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance # 'Private' attribute using name mangling

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self): # Public method to access 'private' balance
        return self.__balance

account = BankAccount("Alice", 1000)
account.deposit(500)
account.withdraw(200)
print(f"Current balance: ${account.get_balance()}")

# Attempting to directly access '__balance' (Python 'name mangles' it)
# print(account.__balance) # This would raise an AttributeError
print(f"Direct access (mangled name): {account._BankAccount__balance}") # Still accessible, but discouraged
```

### 7. Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the essential features of an object. Users interact with simplified interfaces rather than intricate underlying code. While Python doesn't have explicit abstract classes like some other languages, you can achieve abstraction using abstract base classes (ABCs) from the `abc` module.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC): # Abstract Base Class
    @abstractmethod
    def drive(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def drive(self):
        return "Car is driving."

    def stop(self):
        return "Car has stopped."

class Motorcycle(Vehicle):
    def drive(self):
        return "Motorcycle is riding."

    def stop(self):
        return "Motorcycle has stopped."

# You cannot instantiate an abstract class directly
# vehicle = Vehicle() # This would raise a TypeError

my_car = Car()
my_motorcycle = Motorcycle()

print(my_car.drive())
print(my_motorcycle.stop())
```
